```python
import pandas as pd
```

```python
df = pd.read_csv('../tmp/result.csv')
```

```python
# 障害競争を除外
except_obstacle = (df['トラックコード'] == 0) | (df['トラックコード'] == 1) | (df['トラックコード'] == 8)
# 正常発走以外を除外
except_exception = df['異常コード'] == 0
# 馬券内のみ
#arrival_cond = (df['確定着順'] > 0) & (df['確定着順'] <=3)
arrival_cond = df['確定着順'] == 1
# 良馬場のみ
ground_condition_cond = (df['馬場状態'] == '良')

cond = except_obstacle & except_exception & arrival_cond & ground_condition_cond
df = df[cond]

# 開催地、芝orダート、距離で分類
grp_cols = [
    "場所",
    "芝・ダ",
    "距離",
    #"馬場状態",
    #"クラスコード",
]
# 集計
summary = (
    df.groupby(grp_cols)
    .agg(
        平均秒=("走破タイム", "mean"),
        中央値秒=("走破タイム", "median"),
        min_秒=("走破タイム", "min"),
        max_秒=("走破タイム", "max"),
        偏差秒=("走破タイム", "std"),
        サンプル数=("走破タイム", "size"),
    )
    .reset_index()
    .sort_values("サンプル数", ascending=False)
    .reset_index(drop=True))
```
