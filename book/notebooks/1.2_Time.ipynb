# 基準タイム

```python
import pandas as pd
```

```python
df = pd.read_csv("../tmp/result.csv")
```

```python
# --- 1着馬だけに絞る ----------------------------------------------
winners = df.loc[df["確定着順"] == 1]

grp_cols = [
    "場所",
    "芝・ダ",
    "距離",
    "馬場状態",
    "クラスコード",
]

# --- 3. 集計 ----------------------------------------------------------
summary = (
    winners.groupby(grp_cols)
    .agg(
        平均秒=("走破タイム", "mean"),
        中央値秒=("走破タイム", "median"),
        min_秒=("走破タイム", "min"),
        max_秒=("走破タイム", "max"),
        偏差秒=("走破タイム", "std"),
        サンプル数=("走破タイム", "size"),
    )
    .reset_index()
    .sort_values("サンプル数", ascending=False)
    .reset_index(drop=True)
    # .sort_values(['場所','芝・ダ','距離','クラスコード','馬場状態'])
)
# --- 4. 結果を確認 ----------------------------------------------------
summary
```
